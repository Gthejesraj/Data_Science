```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y,c='r',alpha =0.9,marker='X')
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
plt.title("This is my x vs y")
plt.grid()
```

```python
x = np.linspace(1,10,100)
y = np.sin(x)
```

```python
 x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(8,2))
plt.plot(x,y,"--r")
plt.xlabel("x data")
plt.ylabel("y data")
plt.title("sin wave graph")
plt.grid()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z,alpha=.8)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```
